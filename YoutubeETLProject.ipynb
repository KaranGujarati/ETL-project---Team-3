```python
# Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Sources
ca_trend_file = 'CAvideos.csv'
url1 = 'https://en.wikipedia.org/wiki/List_of_most-subscribed_YouTube_channels'
url2 = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_channels'
url3 = 'https://techpostplus.com/2019/04/26/youtube-video-categories-list-faqs-and-solutions/'
```

```python
# Read csv
ca_trend_df = pd.read_csv(ca_trend_file)
```

```python
# Rename columns
ca_trend_df = ca_trend_df.drop(columns = ['comments_disabled', 'ratings_disabled', 'video_error_or_removed'])
```

```python
# Rename and check out
trending_youtube = ca_trend_df
trending_youtube.head()
```

```python
# Read html from web
most_sub_df = pd.read_html(url1)[0]
```

```python
# Drop references
most_sub_df['Channel'] = most_sub_df['Channel'].str.split('[').str[0]
most_sub_df['Network[B]'] = most_sub_df['Network[B]'].str.split('[').str[0]
most_sub_df['Primarylanguage(s)'] = most_sub_df['Primarylanguage(s)'].str.split('[').str[0]

# Rename columns
most_sub_df = most_sub_df.rename(columns={'Rank':'rank', 
                                          'Channel':'channel_title', 
                                          'Subscribers[2](millions)':'subscribers_in_millions',
                                          'Network[B]':'network', 
                                          'Primarylanguage(s)':'language', 
                                          'Contentcategory[2]':'category'})

# Drop last line that states when data was last updated
most_sub_df = most_sub_df.drop(most_sub_df.index[50])
```

```python
# Rename and check out
most_subscribed_channels = most_sub_df
most_subscribed_channels.head()
```

```python
# Read html 
most_views_df = pd.read_html(url2)[0]
```

```python
# Drop references and periods after ranks
most_views_df['Channel name'] = most_views_df['Channel name'].str.split('[').str[0]
most_views_df['Rank'] = most_views_df['Rank'].str.split('.').str[0]

# Rename columns
most_views_df = most_views_df.rename(columns={'Rank':'rank', 
                                          'Channel name':'channel_title', 
                                          'Network':'network',
                                          'Views[1](billions)':'views_in_billions',
                                          'Contentcategory[1]':'category',
                                          'Country':'country',
                                          'Primary language(s)':'language'})
most_views_df = most_views_df.drop(most_views_df.index[50])
```

```python
# Rename and check out
most_viewed_youtubers = most_views_df
most_viewed_youtubers.head()
```

```python
# Read html from web
categories_df = pd.read_html(url3)[0]
```

```python
# Rename columns
categories_df = categories_df.rename(columns={'ID':'category_id', 
                                          'Category name':'category', 
                                          'Category detail':'detail'})
```

```python
# Rename and check out
categories = categories_df.head()
categories.head()
```

```python
# Merge
channel_list = ca_trend_df.merge(most_sub_df, on='channel_title', how='outer')
```

```python
# Merge
channel_list = channel_list.merge(most_views_df, on='channel_title', how='outer')
```

```python
# Create channel series that lists all channels in our data
channel_list = channel_list['channel_title']
```

```python
# Drop all duplicates
channel_list = channel_list.drop_duplicates()
```

```python
# Create data frame, reset index and rename columns
channel_df = pd.DataFrame(channel_list)
channel_df = channel_df.reset_index()
channel_df = channel_df.rename(columns={'index':'channel_no'})
channels = channel_df
channels.head()
```

```python
# Connect to database
rds_connection_string = "postgres:root123@localhost:5432/youtube_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Check table names
engine.table_names()
```

```python
# Load tables
channels.to_sql(name='channels', con=engine, if_exists='append', index=False)
categories.to_sql(name='categories', con=engine, if_exists='append', index=False)
trending_youtube.to_sql(name='trending_youtube', con=engine, if_exists='append', index=False)
most_viewed_youtubers.to_sql(name='most_viewed_youtubers', con=engine, if_exists='append', index=False)
most_subscribed_channels.to_sql(name='most_subscribed_channels', con=engine, if_exists='append', index=False)
```

```python
# Check
pd.read_sql_query('select * from channels', con=engine).head()
```
